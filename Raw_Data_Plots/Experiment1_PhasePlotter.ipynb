```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
mpl.rcdefaults()
```

```python
df = pd.read_csv('Experiment_1_Phase.csv')
```

```python
plt.figure(figsize=(12,8))

plt.title('Phase Difference',fontsize=20)
plt.plot(df.iloc[:, 0][0:128], df.iloc[:, 3][0:128], color='r', linestyle='--')
plt.xlabel('Frequency (Hz)',fontsize=16)
plt.ylabel('Phase Difference (Rad)',fontsize=16)
plt.axvline(x = 2.32, color = 'k', linestyle='-.')
plt.grid()
```
