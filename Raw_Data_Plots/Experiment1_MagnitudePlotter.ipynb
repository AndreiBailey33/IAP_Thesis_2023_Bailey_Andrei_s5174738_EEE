```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
```

```python
mpl.rcdefaults()
```

```python
df = pd.read_csv('Experiment_1_Magnitude.csv')
```

```python
plt.figure(figsize=(10,8))

plt.title('Magnitude Response',fontsize=20)
plt.plot(df.iloc[:, 0][0:128], df.iloc[:, 1][0:128], color='r', linestyle='--')
plt.plot(df.iloc[:, 0][0:128], df.iloc[:, 2][0:128], color='k', linestyle='-.')
plt.xlabel('Frequency (Hz)',fontsize=16)
plt.ylabel('Magnitude',fontsize=16)
plt.grid()
plt.legend(['Sender', 'Receiver'],fontsize=12)
```
